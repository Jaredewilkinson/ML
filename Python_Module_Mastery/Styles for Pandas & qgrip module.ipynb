# Styles for Pandas and qgrip module
Hard to beat pandas documentation for anything including the styles section. My goal is to use the boston dataset, put it into a dataframe, play with styles, and then see what qgrip has to offer. 

+ https://pandas.pydata.org/pandas-docs/version/0.18/style.html

```python
#load in dataset
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
import pandas as pd
```

```python
#make pandas dataframe
df = pd.DataFrame(boston['data'], columns = boston['feature_names'])
```

```python
#add another column to dataframe
df['price']= boston['target']
```

```python
#check to see how it looks
hdf = df.head(10)
hdf
```

```python
#highlight the max value of each column
hdf.style.highlight_max()
```

```python
#The higher the value the darker blue the cell will be (heat map).
hdf.style.background_gradient()
```

```python
#bar width is proportional to the value of the cell
hdf.style.bar()
```

```python
# creating own function to apply to dataframe
def yellow_odd(val):
    if val % 2 == 1:
        return 'color: yellow' # will color the number this color.
    return ''
```

```python
hdf.style.applymap(yellow_odd)
```

```python
# creating own function to apply to dataframe
def purple5(val):
    if val > 5:
        return 'background-color: purple'#will color the background purple
    return ''
```

```python
hdf.style.applymap(yellow_odd).applymap(purple5)#chaining together both functions
```

```python
#appling function to only a subset of the data(slicing).
hdf.style.applymap(yellow_odd, subset=['DIS','RAD'])
```

```python
# Another function
def top_50(col):
    top = col > col.quantile(.5) # fifty so its easier to see the effect.
    return ['font-weight: bold' if x else '' for x in top]
   
```

```python
#apply works on columns and rows

hdf.style.apply(top_50)
```

```python
import seaborn as sns
#choose a color map
cm = sns.light_palette("orange", as_cmap=True)

s = hdf.style.background_gradient(cmap=cm)
s
```

```python
 hdf.style.background_gradient(cmap=cm).set_caption('Heat map of Boston Dataset')
```

```python
help(hdf.style.set_caption)
```

```python
hdf.style.bar(subset=['LSTAT', 'price'], color='#d65f5f')
```

```python
import numpy as np
import pandas as pd
import qgrid
randn = np.random.randn
df_types = pd.DataFrame({
    'A' : pd.Series(['2013-01-01', '2013-01-02', '2013-01-03', '2013-01-04',
               '2013-01-05', '2013-01-06', '2013-01-07', '2013-01-08', '2013-01-09'],index=list(range(9)),dtype='datetime64[ns]'),
    'B' : pd.Series(randn(9),index=list(range(9)),dtype='float32'),
    'C' : pd.Categorical(["washington", "adams", "washington", "madison", "lincoln","jefferson", "hamilton", "roosevelt", "kennedy"]),
    'D' : ["foo", "bar", "buzz", "bippity","boppity", "foo", "foo", "bar", "zoo"] })
df_types['E'] = df_types['D'] == 'foo'
qgrid_widget = qgrid.QgridWidget(df=df_types, show_toolbar=True)
qgrid_widget
```

```python
newdf = qgrid_widget.get_changed_df()
```
