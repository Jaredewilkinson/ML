```python
xs = np.linspace(-10,10,100)
ys = np.sinc(xs)
```

```python
df = pd.DataFrame({'x':xs , 'sinc':ys})
```

```python
plt.plot(xs,ys, label = r'$\frac{sin(x)}{x}$')
plt.legend()
```

```python
plt.plot(xs,ys, label = r'$\frac{sin(x)}{x}$')
plt.plot(xs,np.cos(xs), label = r'cos(xs)')
plt.legend()
```

```python
plt.subplot(2,1,1)
plt.plot(xs,np.cos(xs), label = r'cos(xs)')
plt.legend()
plt.subplot(2,1,2)
plt.fill(xs,-xs**2, label =r'${-xs}**2$')
```

```python
df = pd.read_csv('lnkd.csv', parse_dates=['Date'], index_col ='Date')
df.head()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize =(15,6))
df.loc['2016-04', 'Close'].plot(ax= ax1)
df.loc['2016-04','Volume'].plot(ax= ax2)
```

```python
cax, vax = df[['Close','Volume']].plot(subplots = True)
xmin, xmax = cax.get_xlim()
cax.hlines(df['Close'].mean(),xmin, xmax, color ='green')
vax.hlines(df['Volume'].median(),xmin, xmax, color ='green')
```

# Interactive plots using %matplotlib notebook

```python
# can turn on and off with the power button
%matplotlib notebook

cax, vax = df[['Close','Volume']].plot(subplots = True)
xmin, xmax = cax.get_xlim()
cax.hlines(df['Close'].mean(),xmin, xmax, color ='green')
vax.hlines(df['Volume'].median(),xmin, xmax, color ='green')
```

```python
@interact(limit=6)
def plot_lim(limit):
    xs = np.linspace(-limit,limit, 100)
    plt.plot(xs, np.cos(xs), label='cos(x) [{}-{}]'.format(-limit,limit))
```
